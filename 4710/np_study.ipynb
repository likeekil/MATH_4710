```python
import numpy as np
import matplotlib.pyplot as plt

# ========= 累加表格数据（行=累计到该年，列=灾害）=========
disasters = ["Freeze", "Wind", "Flood", "Quake", "Hail"]

# 行标签：去掉 (cum)
row_labels = ["100 y", "50 y", "20 y", "10 y", "5 y"]

# 累加值（单位：$1000's）
table_cum = [
    [431.5, 1049.4,  799.6, 2149.8,  917.9],  # ≤100y
    [292.2,  717.8,  456.4, 1368.5,  865.6],  # ≤50y
    [213.8,  636.0,  305.7, 1115.2,  796.0],  # ≤20y
    [124.6,  555.4,  153.2,  677.2,  192.5],  # ≤10y
    [ 66.4,  174.3,   75.1,  577.9,   32.0],  # ≤5y
]
# ===============================================

cum = np.array(table_cum, dtype=float)
order_small_to_large = [4, 3, 2, 1, 0]  # 从 5→100

# 把累加转为分段值
segments = []
for k, r_idx in enumerate(order_small_to_large):
    if k == 0:
        seg = cum[r_idx]
    else:
        prev = cum[order_small_to_large[k - 1]]
        seg = cum[r_idx] - prev
    seg = np.clip(seg, 0, None)
    segments.append(seg)
segments = np.array(segments)

# 颜色（从下到上：5→100）
colors = ["#A8E6CF", "#DCEDC1", "#FFD3B6", "#FFAAA5", "#FF8C94"]

# ============ 绘图 ============
fig, ax = plt.subplots(figsize=(9, 6))
x = np.arange(len(disasters))
bottom = np.zeros(len(disasters))

for i in range(len(segments)):
    ax.bar(x, segments[i], bottom=bottom, color=colors[i], edgecolor='black')
    bottom += segments[i]

# 标题/坐标轴
ax.set_title("Loss by Disaster", fontsize=18, fontweight='bold', pad=20)
ax.set_ylabel("Loss $1000’s", fontsize=12)
ax.set_ylim(0, 2500)
ax.set_yticks(np.arange(0, 2501, 500))
ax.grid(True, axis="y", linestyle=" ", linewidth=0.5, alpha=0.7)

# 不显示 x 轴标签（灾害名放表格）
ax.set_xticks([])
ax.set_xlabel("")

# ====== 下方表格 ======
tbl = plt.table(cellText=np.round(cum, 1),
                rowLabels=row_labels,
                colLabels=disasters,
                loc='bottom',
                cellLoc='center',
                rowLoc='center')
tbl.scale(1, 1.5)
tbl.auto_set_font_size(False)
tbl.set_fontsize(9)

# ====== 给表格第一列（年份列）上色 ======
color_map = {
    "5 y":   colors[0],
    "10 y":  colors[1],
    "20 y":  colors[2],
    "50 y":  colors[3],
    "100 y": colors[4],
}

for (r, c), cell in tbl.get_celld().items():
    # 行标签列 c == -1；跳过表头 r == -1
    if c == -1 and r != -1:
        text = cell.get_text().get_text().strip()
        if text in color_map:
            cell.set_facecolor(color_map[text])
            cell.set_text_props(weight='bold', color='black')

# 布局调整
plt.subplots_adjust(left=0.1, bottom=0.25, right=0.85, top=0.85)
plt.show()
```

```python
# -*- coding: utf-8 -*-
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

# ========= 累加表格数据（行=累计到该年，列=灾害）=========
disasters = ["Freeze", "Wind", "Flood", "Quake", "Hail"]
row_labels = ["100 y", "50 y", "20 y", "10 y", "5 y"]

# 累加值（单位：$1000's）
table_cum = [
    [431.5, 1049.4,  799.6, 2149.8,  917.9],
    [292.2,  717.8,  456.4, 1368.5,  865.6],
    [213.8,  636.0,  305.7, 1115.2,  796.0],
    [124.6,  555.4,  153.2,  677.2,  192.5],
    [ 66.4,  174.3,   75.1,  577.9,   32.0],
]
cum = np.array(table_cum, dtype=float)   # (5 periods × 5 disasters)

# —— 转换累加为分段值（用于堆叠：5→100）
order_small_to_large = [4, 3, 2, 1, 0]  # 5,10,20,50,100 的行索引
segments = []
for k, r_idx in enumerate(order_small_to_large):
    if k == 0:
        seg = cum[r_idx]
    else:
        seg = cum[r_idx] - cum[order_small_to_large[k - 1]]
    segments.append(np.clip(seg, 0, None))
segments = np.array(segments)

# —— 配色定义
periods_small_to_large = [5, 10, 20, 50, 100]
period_colors = ["#A8E6CF", "#DCEDC1", "#FFD3B6", "#FFAAA5", "#FF8C94"]  # 5,10,20,50,100
period_color_map = dict(zip([f"{p} y" for p in periods_small_to_large], period_colors))

disaster_colors = ["#6EC1E4", "#7BDCB5", "#FAD27E", "#F3A683", "#C39BD3"]  # Freeze, Wind, Flood, Quake, Hail

# ======== 布局 2×3 ========
fig = plt.figure(figsize=(14, 9))
gs = GridSpec(2, 3, figure=fig, width_ratios=[2, 2, 1], height_ratios=[1.3, 1])
gs.update(hspace=0.65)  # 增加上下间距

ax_bar  = fig.add_subplot(gs[0, 0:2])  # 左上：堆叠柱
ax_line = fig.add_subplot(gs[1, 0:2])  # 左下：折线（x=disaster, 每条线=时间段）
ax_avg  = fig.add_subplot(gs[:, 2])    # 右侧：Average Loss（星形散点）

# ======== (1) 堆叠柱 + 表格 ========
x = np.arange(len(disasters))
bottom = np.zeros(len(disasters))
for i in range(len(segments)):  # i: 0..4 -> 5→100
    ax_bar.bar(x, segments[i], bottom=bottom, color=period_colors[i], edgecolor='white')
    bottom += segments[i]

ax_bar.set_title("Loss by Disaster (Stacked)", fontsize=16, fontweight='bold', pad=12)
ax_bar.set_ylabel("Loss ×1000’s", fontsize=11)
ax_bar.set_ylim(0, 2500)
ax_bar.set_yticks(np.arange(0, 2501, 500))
ax_bar.grid(True, axis="y", linestyle="--", linewidth=0.5, alpha=0.7)
ax_bar.set_xticks([])
ax_bar.set_xlabel("")


# 给年份列 & 列头上色
for (r, c), cell in tbl.get_celld().items():
    if c == -1 and r != -1:  # 年份列
        label = cell.get_text().get_text().strip()
        if label in period_color_map:
            cell.set_facecolor(period_color_map[label])
            cell.set_text_props(weight='bold', color='black')
    if r == -1 and 0 <= c < len(disasters):  # 列头
        cell.set_facecolor(disaster_colors[c])
        cell.set_text_props(weight='bold', color='black')

# 调整表格位置
pos = ax_bar.get_position()
ax_bar.set_position([pos.x0, pos.y0 + 0.05, pos.width, pos.height - 0.05])

# ======== (2) 折线图：x = Disaster, y = Loss，每条线代表一个时间段 ========
cum_small_to_large = cum[order_small_to_large, :]  # (5 periods × 5 disasters)

for i, period in enumerate(periods_small_to_large):
    ax_line.plot(disasters, cum_small_to_large[i],
                 marker='o', linewidth=2.5, color=period_colors[i],
                 label=f"{period} y")

ax_line.set_title("Cumulative Loss by Disaster (each Return Period)", fontsize=14, pad=8)
ax_line.set_xlabel("Disaster Type")
ax_line.set_ylabel("Cumulative Loss ×1000’s")
ax_line.grid(True, linestyle="--", linewidth=0.5, alpha=0.7)
ax_line.set_ylim(0, np.max(cum_small_to_large) * 1.15)
ax_line.legend(title="Return Period", ncol=3, frameon=False, fontsize=9, loc="upper left")

# ======== (3) 右侧：Average Loss —— 按时间段，星形散点 ========
avg_by_period = cum_small_to_large.mean(axis=1)
x_ticks = [f"{p} y" for p in periods_small_to_large]
x_pos = np.arange(len(periods_small_to_large))

for i, (xp, val) in enumerate(zip(x_pos, avg_by_period)):
    ax_avg.scatter(xp, val, marker='*', s=350, color=period_colors[i], edgecolors='black', linewidths=0.8)

ax_avg.set_title("Average Cumulative Loss\n(by Return Period)", fontsize=14)
ax_avg.set_ylabel("Average Loss ×1000’s")
ax_avg.set_xticks(x_pos)
ax_avg.set_xticklabels(x_ticks)
ax_avg.grid(True, axis="y", linestyle="--", linewidth=0.5, alpha=0.7)
ax_avg.set_ylim(0, avg_by_period.max() * 1.25)

plt.tight_layout()
plt.show()
```
